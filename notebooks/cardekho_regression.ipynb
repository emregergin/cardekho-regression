```python
X = df.drop("selling_price",axis=1)
y = df["selling_price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=15)
```

```python
cat_cols = [col for col in X_train.columns if X_train[col].dtype == "O"]
```

```python
cat_cols
```

```python
df[cat_cols].nunique()
```

```python
onehot_cols = ["seller_type","fuel_type","transmission_type"]
freq_cols = ["car_name","brand","model"]
```

```python
for col in freq_cols:
    freq = X_train[col].value_counts() / len(X_train)
    X_train[col+"_freq"] = X_train[col].map(freq)
    X_test[col+"_freq"] = X_test[col].map(freq)
    mean_freq = freq.mean()
    X_test[col + "_freq"] = X_test[col + "_freq"].fillna(mean_freq)
```

```python
X_train[X_train["car_name_freq"].isnull()]
```

```python
X_train[X_train["brand_freq"].isnull()]
```

```python
X_train[X_train["model_freq"].isnull()]
```

```python
X_train.drop(["car_name","brand","model"],axis=1,inplace=True)
X_test.drop(["car_name","brand","model"],axis=1,inplace=True)
```

```python
print(X_train.head())
print(X_test.head())
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(
    [("one_hot_encoder",OneHotEncoder(drop="first",handle_unknown="ignore"),onehot_cols)],
    remainder="passthrough"
)
```

```python
X_train = transformer.fit_transform(X_train)
X_test = transformer.transform(X_test)
```

```python
encoded_cols = transformer.get_feature_names_out()
encoded_cols
```

```python
X_train = pd.DataFrame(data=X_train,columns=encoded_cols)
X_test = pd.DataFrame(data=X_test,columns=encoded_cols)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
X_train.info()
```

```python
X_test.info()
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import RandomizedSearchCV
```

```python
model = AdaBoostRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
params = {
    "n_estimators": [50,80,100,120],
    "learning_rate": [0.001,0.1,1,5],
    "loss": ["linear","square","exponential"]
}
```

```python
rscv = RandomizedSearchCV(estimator=AdaBoostRegressor(),param_distributions=params,scoring="r2",n_jobs=-1,cv=5)
```

```python
rscv.fit(X_train,y_train)
```

```python
rscv.best_params_
```

```python
y_pred = rscv.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
new_params = {
    "estimator__max_depth": [3,4,5,6,7,8,9,10],
    "n_estimators": [50,80,100,120],
    "learning_rate": [0.001,0.1,1,5],
    "loss": ["linear","square","exponential"]
}
```

```python
rscv_new = RandomizedSearchCV(estimator=AdaBoostRegressor(DecisionTreeRegressor()),param_distributions=new_params,scoring="r2",cv=5,n_jobs=-1)
```

```python
rscv_new.fit(X_train,y_train)
```

```python
rscv_new.best_params_
```

```python
y_pred_new = rscv_new.predict(X_test)
```

```python
print(r2_score(y_test,y_pred_new))
print(mean_absolute_error(y_test,y_pred_new))
print(mean_squared_error(y_test,y_pred_new))
```
